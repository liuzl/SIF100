```python
from mpmath import mp
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
for n_digits in [100, 10000, 100000, 1000000]:
    mp.dps = n_digits
    pi = [int(x) for x in list(str(mp.pi)[2:])]
    df = pd.DataFrame(pi, columns=['numbers'])
    df.plot(kind='hist', xticks=range(10), grid=True)
```
